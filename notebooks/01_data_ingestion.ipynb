```python

# Day 1: Data Ingestion Script
# Loads all CSV datasets, displays info, and validates data quality

import pandas as pd
from pathlib import Path
```

```python
# Define Paths
RAW_DATA_PATH = Path(r"G:\bluestock_mf_capstone\data\raw")
PROCESSED_DATA_PATH = Path(r"G:\bluestock_mf_capstone\data\processed")

# Create directories if they don't exist
RAW_DATA_PATH.mkdir(parents=True, exist_ok=True)
PROCESSED_DATA_PATH.mkdir(parents=True, exist_ok=True)
```

```python
def load_and_inspect_csv(file_path, file_name):
    """
    Load a CSV file and print its shape, dtypes, and head
    """
    print("\n" + "=" * 80)
    print(f"📊 ANALYZING: {file_name}")
    print("=" * 80)

    try:
        # Load CSV
        df = pd.read_csv(file_path)

        # Shape
        print(f"\n📏 Shape: {df.shape[0]} rows × {df.shape[1]} columns")

        # Data Types
        print("\n📋 Data Types:")
        print(df.dtypes.to_string())

        # Sample Rows
        print("\n👀 First 5 Rows:")
        print(df.head())

        # Data Quality Checks
        print("\n⚠️ Data Quality Check:")

        # Missing values
        missing_cols = df.columns[df.isnull().any()].tolist()

        if missing_cols:
            print("   Missing values found:")
            for col in missing_cols:
                null_count = df[col].isnull().sum()
                null_pct = (null_count / len(df)) * 100
                print(f"   • {col}: {null_count} missing ({null_pct:.2f}%)")
        else:
            print("   ✅ No missing values")

        # Duplicates
        duplicates = df.duplicated().sum()

        if duplicates > 0:
            print(f"   ⚠️ Duplicate rows: {duplicates}")
        else:
            print("   ✅ No duplicate rows")

        # Empty strings
        str_cols = df.select_dtypes(include=["object"]).columns

        for col in str_cols:
            empty_strings = (df[col].astype(str).str.strip() == "").sum()

            if empty_strings > 0:
                print(f"   ⚠️ {col}: {empty_strings} empty strings")

        return df

    except Exception as e:
        print(f"❌ Error loading {file_name}")
        print(f"   {e}")
        return None
```

```python
def main():
    """
    Main function
    """
    print("\n🚀 STARTING DATA INGESTION PROCESS")
    print("=" * 80)

    
    # Debug Path Information
    
    print("\n📂 PATH CHECK")
    print("-" * 40)
    print("Raw Path:", RAW_DATA_PATH)
    print("Exists:", RAW_DATA_PATH.exists())
    print("Is Directory:", RAW_DATA_PATH.is_dir())

    if RAW_DATA_PATH.exists():
        print("\n📄 Files in Raw Folder:")

        try:
            for item in RAW_DATA_PATH.iterdir():
                print("   ", item.name)
        except Exception as e:
            print("Error reading folder:", e)

    
    # Find CSV Files
    
    csv_files = [
        f for f in RAW_DATA_PATH.iterdir()
        if f.is_file() and f.suffix.lower() == ".csv"
    ]

    if not csv_files:
        print("\n❌ No CSV files found!")
        print(f"Expected location: {RAW_DATA_PATH}")
        return

    print(f"\n✅ Found {len(csv_files)} CSV file(s)")

    
    # Load Files
    
    dataframes = {}

    for csv_file in csv_files:
        df = load_and_inspect_csv(csv_file, csv_file.name)

        if df is not None:
            dataframes[csv_file.stem] = df
            # Summary
    
    print("\n" + "=" * 80)
    print("📈 INGESTION SUMMARY")
    print("=" * 80)

    print(f"✅ Successfully Loaded: {len(dataframes)}")
    print(f"❌ Failed: {len(csv_files) - len(dataframes)}")

    if dataframes:
        metadata = {
            "files_processed": len(dataframes),
            "file_names": ", ".join(dataframes.keys()),
            "total_rows": sum(len(df) for df in dataframes.values()),
            "total_columns": sum(len(df.columns) for df in dataframes.values())
        }

        metadata_df = pd.DataFrame([metadata])

        metadata_file = PROCESSED_DATA_PATH / "ingestion_metadata.csv"

        metadata_df.to_csv(metadata_file, index=False)

        print(f"\n💾 Metadata saved to:")
        print(metadata_file)

    print("\n✨ Data ingestion complete!")


if __name__ == "__main__":
    main() 
```
